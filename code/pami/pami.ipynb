```python
import pandas as pd
from ast import literal_eval
```

```python
retail_df = pd.read_csv('../../data/retail-transactions/Retail_Transactions_Dataset.csv')
order_products_train_df = pd.read_csv('../../data/instacart-market-basket-analysis/order_products_train.csv')
order_products_prior_df = pd.read_csv('../../data/instacart-market-basket-analysis/order_products_prior.csv')
products_df = pd.read_csv('../../data/instacart-market-basket-analysis/products.csv')

# Displaying the first few rows of each dataset
print(retail_df.head())
print(order_products_train_df.head())
print(order_products_prior_df.head())
print(products_df.head())
```

```python
print(order_products_train_df.head())
print(order_products_prior_df.head())
print(products_df.head())
```

Data Cleaning

```python
# Remove duplicates
retail_df = retail_df.drop_duplicates()
order_products_train_df = order_products_train_df.drop_duplicates()
order_products_prior_df = order_products_prior_df.drop_duplicates()
products_df = products_df.drop_duplicates()

#Remove missing values
retail_df = retail_df.dropna()
order_products_train_df = order_products_train_df.dropna()
order_products_prior_df = order_products_prior_df.dropna()
products_df = products_df.dropna()
```

```python
order_products_train_df = order_products_train_df.drop_duplicates()
order_products_prior_df = order_products_prior_df.drop_duplicates()
products_df = products_df.drop_duplicates()

#Remove missing values
retail_df = retail_df.dropna()
order_products_train_df = order_products_train_df.dropna()
order_products_prior_df = order_products_prior_df.dropna()
products_df = products_df.dropna()
```

```python
print(len(retail_df))
print(len(order_products_train_df))
print(len(order_products_prior_df))
print(len(products_df))
```

Data Integration

```python
# Join order_products_train with products

# instacart_df = pd.merge(order_products_train_df, products_df, on='product_id', how='inner')

instacart_df = pd.concat([order_products_train_df, order_products_prior_df])
instacart_df = pd.merge(instacart_df, products_df, on='product_id', how='inner')

print(instacart_df.head())
```

Data Transformation

```python
retail_df['Product'] = retail_df['Product'].apply(literal_eval)
retail_df = retail_df.explode('Product', ignore_index=True)
```

```python
# Rename attributes biar sama
retail_df = retail_df.rename(columns={'Product': 'product_name', 'Transaction_ID': 'order_id'})

retail_df = retail_df[['product_name', 'order_id']]  
instacart_df = instacart_df[['product_name', 'order_id']] 
```

```python
# instacart_df = instacart_df.sample(frac=0.01, random_state=42)
instacart_df
```

```python
# retail_df = retail_df.sample(frac=0.01, random_state=42)
retail_df
```

Merge Stack Verically

```python
combined_df = pd.concat([retail_df, instacart_df], ignore_index=True)
combined_df
```

```python
# Bikin kolom product ID baru (terpisah dari punya instacart)
combined_df['product_id'] = 'P' + (combined_df['product_name'].astype('category').cat.codes + 1).astype(str)
```

```python
combined_df
```

```python
# Group product_name by order_id to form transactions
# transactions = combined_df.groupby('order_id')['product_id'].apply(list).tolist() # by ID
transactions = combined_df.groupby('order_id')['product_name'].apply(list).tolist() # by name

# Display the first few transactions
print(len(transactions))
```

```python
# Filepath for the output file
output_file = "../../data/transformed/transactions.csv"

# Open the file in write mode
with open(output_file, "w", encoding="utf-8") as file:
    for transaction in transactions:
        # Join items in each transaction with a delimiter (e.g., space or comma)
        line = "\t".join(transaction)  # Use "," for comma-separated values if needed
        file.write(line + "\n")  # Write each transaction on a new line

print(f"Transactions exported to {output_file}")
```

# PAMI

```python
minSup = 5000
sep = '\t'
```

```python
minConf = 0.01
minLift = 0.01
maxTS = 99935
```

# FPGrowth

```python
from PAMI.frequentPattern.basic import FPGrowth as alg
```

```python
obj = alg.FPGrowth(output_file, minSup, sep=sep)
```

```python
obj.mine()
```

```python
# obj.save('frequentPatterns_FPGrowth.txt', seperator=sep)
obj.save('../../output/pami/fp_growth/frequentPatterns_FPGrowth.csv')
```

```python
frequentPatternsDataFrame = obj.getPatternsAsDataFrame()

frequentPatternsDataFrame
```

```python
from PAMI.AssociationRules.basic import confidence as alg
obj = alg.confidence(frequentPatternsDataFrame, minConf, sep)
obj.mine()
obj.printResults()
obj.save("../../output/pami/fp_growth/associationRulesconfidence_FPGrowth.csv")
obj.getAssociationRulesAsDataFrame()
```

```python
from PAMI.AssociationRules.basic import lift as alg
obj = alg.lift(frequentPatternsDataFrame, minLift, sep)
obj.mine()
obj.printResults()
obj.save("../../output/pami/fp_growth/associationRuleslift_FPGrowth.csv")
test = obj.getPatternsAsDataFrame()
test
```

```python
from PAMI.AssociationRules.basic import leverage as alg
obj = alg.leverage(frequentPatternsDataFrame, minLift, sep, maxTS)
obj.mine()
obj.printResults()
obj.save("../../output/pami/fp_growth/associationRulesleverage_FPGrowth.csv")
obj.getAssociationRulesAsDataFrame()
```
